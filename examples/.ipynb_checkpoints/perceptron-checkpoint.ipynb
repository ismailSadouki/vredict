We will apply the perceptron algorithm on the Iris dataset.

Note that we will only consider two flower classes, setosa and versicolor, from the Iris dataset for practical reasons—remember, the perceptron is a binary classifier. However, the perceptron algorithm can be extended to multi-class classification—for example, <mark>the one-versus-all (OvA)</mark> technique.


Let's start by loading the data:

```python
from sklearn import datasets
import matplotlib.pyplot as plt
iris = datasets.load_iris()
X = iris.data[:, [2,3]]
y = iris.target
```

```python

# Only keep classes 0 and 1
X = X[y < 2]
y = y[y < 2]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0, 0], X[y == 0, 1],
            color='red', marker='o', label='Setosa (0)')
plt.scatter(X[y == 1, 0], X[y == 1, 1],
            color='blue', marker='x', label='Versicolor (1)')

plt.xlabel('Petal length [cm]')
plt.ylabel('Petal width [cm]')
plt.title('Iris Classes: Setosa vs Versicolor')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

the distribution of flower examples in the Iris dataset along the two feature axes: petallength and sepal length (measured in centimeters). In this two-dimensional feature subspace, we can see that a linear decision boundary should be sufficient to separate setosa from versicolor flowers. Thus, a linear classifier such as the perceptron should be able to classify the flowers in this dataset perfectly.



train our perceptron algorithm on the Iris data subset that we just extracted.

```python
import sys
import os
sys.path.append(os.path.abspath(".."))
from vredict.perceptron import Perceptron

ppn = Perceptron(eta=0.1, n_iter=10)
ppn.fit(X, y)
```

 plot the misclassification error for each epoch to check whether the algorithm converged

```python
import matplotlib.pyplot as plt
plt.plot(range(1, len(ppn.errors_)+1), ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')
```

The model found a decision boundary by epoch 3, and then made no more updates after that — this is the expected behavior when the dataset is linearly separable, which the two-class Iris subset is.

Let's visualize the Perceptron's decision regions

```python
from vredict.utils import plot_decision_regions

plot_decision_regions(X, y, classifier=ppn)
plt.xlabel('Petal length [cm]')
plt.ylabel('Petal width [cm]')
plt.legend(loc='upper left')
plt.title('Perceptron - Decision Boundary')
plt.show()
```
